# Descriptive Statistics 

### **Easy Level**

**Q1: Understanding Central Tendency (Easy):  A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12]. What is the most representative value of their weekly sales, and why?**
- Most representative value = 12.43 dozens (Mean)
- Reason:
  - Because the dataset has no outliers and is fairly balanced, the mean gives the most accurate central tendency by considering all values.

<br>

**Q2: Mean in Real Life (Easy): A teacher records the marks of her students in a short quiz: [12, 15, 14, 16, 18, 20, 19]. What is the mean score, and what does it tell us about the class’s performance?**
-   Mean score = (12+15+14+16+18+20+19) / 7 = 16.28

- A mean of 16.28 out of 20 indicates:
  - The class performed above average.
  - Most students scored in the mid-to-high range.
  - There are no extremely low marks pulling the average down.
  - Overall, the class shows good understanding of the quiz topics.

<br>

**Q3: Mode in Real Life (Easy):  A store records the shoe sizes sold in one day: [7, 8, 9, 8, 8, 10, 7, 9]. What is the mode, and why is this information useful for the store manager?**
- Mode → Most frequently sold size = 8 (appears 3 times)
- Reason:
  - Knowing the mode helps the store manager understand which shoe size is most in demand.

### **Medium Level**

**Q4: Median in Real Life (Medium):  A car dealer notes the prices of used cars: [$8,000, $9,500, $10,200, $11,000, $50,000]. Why is the median a better measure than the mean in this case? Calculate the median.**
- Since there are 5 values, the median is the 3rd value in the sorted list: ($8,000, $9,500, $10,200, $11,000, $50,000)
  - Median = $10,200
- The median a better measure:
  - Because the dataset has an outlier — $50,000 and the median ignores extreme values and represents the central tendency better. Most cars are priced between $8,000–$11,000, so the median price of $10,200 is a more realistic measure.

<br>

**Q5: Dispersion Introduction (Medium): A student times how long it takes to finish a puzzle each day: [25, 30, 27, 35, 40]. What does the range tell us about the variation in the student’s puzzle-solving time?**
- Range = 40 - 25 = 15
- A range of 15 minutes means that the student’s completion times vary by up to 15 minutes between the fastest day (25 min) and the slowest day (40 min).

<br>

**Q6: Range in Action (Medium): A farmer records the weekly weight of harvested apples (kg): [100, 105, 98, 110, 120]. Find the range. How can this help the farmer in planning his packaging?**
- Range = 120 − 98 = 22 kg
- This helps the farmer know how much the harvest varies each week, so he can plan enough packaging and storage for weeks with higher weights.

### **Hard Level**

**Q8: Standard Deviation in Context (Hard): A finance student compares the daily price fluctuations of two cryptocurrencies.**
**Coin A: standard deviation = $30**
**Coin B: standard deviation = $120**
**Which coin is riskier to invest in, and why?**
- Coin B is riskier to invest in.
- Because: 
  - Coin A (SD = $30) → Smaller fluctuations → More stable.
  - Coin B (SD = $120) → Larger fluctuations → Highly unstable.
- Coin B’s standard deviation is 4 times larger, its price swings much more widely. Greater price variability means greater uncertainty and higher risk, so Coin B is riskier.

**Q9: Combining Measures (Hard): A family records their monthly electricity usage (in kWh): [400, 420, 390, 450, 410].Find the mean and standard deviation. What do these values together tell you about the family’s energy use pattern?**

- Data: [400, 420, 390, 450, 410]

- **Mean** = (400 + 420 + 390 + 450 + 410) / 5 = **414**

- **Standard Deviation**

| Score | (x − μ)² |
|-------|----------|
| 400 | 196 |
| 410 | 16 |
| 420 | 36 |
| 450 | 1296 |
| 390 | 576 |

  - Sum of squares = **2120**
  - sigma^2 = 2120 / 5 = **424**
  - sigma = sqrt{8.4825} ≈ **20.591**

**Standard Deviation ≈ 20.591**


- **Final Summary**

| Measure | Value |
|--------|--------|
| **Mean** | **19.625** |
| **Standard Deviation** | **≈ 20.591** |


- Insights About Performance:
  - The family’s energy usage is fairly consistent, with occasional higher or lower months (e.g., 450 kWh in one month). The SD tells us how much their usage typically deviates from the average.

**Q10: Practical Application (Hard): A basketball player’s points in 8 games are recorded: [15, 18, 20, 22, 25, 17, 19, 21]. Find the mean, median, mode, range, and standard  eviation. What insights can these measures provide about the player’s scoring performance?**

- Data: [15, 17, 18, 19, 20, 21, 22, 25]

- **Mean** = (15 + 17 + 18 + 19 + 20 + 21 + 22 + 25) / 8 = **19.625**

- **Median** = Middle two values: 19 and 20: (19 + 20) / (2) = **19.5**

- **Mode**: All values occur once. = **None**

- **Range** = 25 - 15 = **10**

- **Standard Deviation**

| Score | (x − μ)² |
|-------|----------|
| 15 | 21.39 |
| 17 | 6.89 |
| 18 | 2.64 |
| 19 | 0.39 |
| 20 | 0.14 |
| 21 | 1.89 |
| 22 | 5.64 |
| 25 | 28.89 |

  - Sum of squares = **67.86**
  - sigma^2 = frac{67.86}{8} = **8.4825**
  - sigma = sqrt{8.4825} ≈ **2.91**

**Standard Deviation ≈ 2.91**


- **Final Summary**

| Measure | Value |
|--------|--------|
| **Mean** | **19.625** |
| **Median** | **19.5** |
| **Mode** | **None** |
| **Range** | **10** |
| **Standard Deviation** | **≈ 2.91** |


- Insights About Performance

  - Stable Scoring: The low SD (~2.91) shows **consistent performance**.
  - No Outliers: Scores vary only within a **10-point range**.
  - Symmetrical Distribution: Mean ≈ Median → **no skewness**, balanced scoring.
  - No Mode: No repeated score → slight variation every game.
