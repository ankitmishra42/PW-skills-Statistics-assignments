**Q4: The population of a school is divided into 60% boys and 40% girls. If you want equal representation of both genders in the sample, which method should you use: Simple Random Sampling or Stratified Sampling? Why?**
- I will use Stratified Sampling:<br>
  - Because it allows me to divide the population into two strata boys and girls and then select equal samples from each group. This ensures equal representation, which Simple Random Sampling cannot guarantee.<br>

- Why Stratified Sampling works:<br>
  - In Stratified Sampling, I divide the population into subgroups (strata) (here: boys and girls).<br>
  - Then I sample equally (or proportionally, depending on the requirement) from each strata.<br>
  - If I want a 50-50 gender representation, I can deliberately select equal numbers from each strata.<br>
